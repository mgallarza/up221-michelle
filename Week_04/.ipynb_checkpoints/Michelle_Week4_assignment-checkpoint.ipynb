Since the data has separate columns for latitude and longitude, the data can be converted to a geodataframe.

Lat and long coordinates system (GCS) have a crs code of epsg:4326.

```python
gdf = gpd.GeoDataFrame(df, 
                       crs='epsg:4326',
                       geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
```
