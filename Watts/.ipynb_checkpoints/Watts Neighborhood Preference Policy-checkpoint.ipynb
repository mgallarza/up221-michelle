```python
df = pd.read_csv('data/Los_Angeles_County_CVA_Social_Vulnerability_Index.csv')
```

```python
df.info()
```

```python
# try it yourself. Query the dataframe for other properties of interest
df.query("CSA_Label == 'Los Angeles - Watts'")
```

```python
df_trimmed = df.query("CSA_Label == 'Los Angeles - Watts'").copy()
df_trimmed
```

```python
#sort the columns in descending order
df_sorted = df_trimmed.sort_values(by='Rent_Burden',ascending = False)
```

```python
df_sorted.plot.bar(x='Census_Tract',
                     y='Rent_Burden',
                     title='% Households Rent Burdened in Watts (By Census Tract)',
                     color='purple')
```
