## Data Plotting with Single Variables

Now that my data is trimmed for the variables I want and sorted for City of Los Angeles neighborhoods, I can begin creating some graphs. I will start with some single-variable practice charts to idenitfy the top 10 neighborhoods/tracts for Households Without Vehicle Acccess, Rent Burdened Households, Mobile Homes, and Disability (based on percentage).

### Households Without Vehicle Access
This chart reflects a high concentration of households in the inner city/downtown area. While DTLA may not experience as may hot days as other areas of the city, without access to a private vehicle these residents are more likely to be exposed to extreme heat outside when heat waves do occur.

```python
#sort the columns in descending order
lac_sorted = lac.sort_values(by='Households_Without_Vehicle_Acce',ascending = False)
```

```python
lac_sorted.head(10).plot.bar(x='CSA_Label',
                     y='Households_Without_Vehicle_Acce',
                     title='Top 10 Census Tracts for Households Without Vehicle Access in Los Angeles City',
                     color='purple')
```

This test chart with two variables attempts to look at the connection between households without cars and older adults (age 65 and over).
